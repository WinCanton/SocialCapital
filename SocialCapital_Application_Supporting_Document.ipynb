## 3. Relative Strength Trading Strategy

The trade-setup mentioned in the previous section is based on a strategy that I refer to here as the Relative Strength trading strategy. As the name implies, it is a momentum strategy that utilises information provided by the market itself on its relative-strength. This information is derived from a formula introduced by Welles Wilder in 1978, commonly known as the RSI formula. This formula calculates a value that represents a ratio of the average recent time period 'gains' to the average recent time period price 'losses' calculated over a number of time periods. This information represents the internal strength of a market.

The Relative Strength trading strategy then utilises the market's internal strength as a means to discover the following market properties which are very beneficial in trading:

1. The current trend of the market
2. The best entry/exit prices for a trade-setup
3. Price levels where a pullback will likely end
4. Determine the dominant market timeframe
5. When a higher-timeframe traders overpowering a lower-timeframe ones
6. The highly probable price targets to be reached

Following from the fact that the internal strength information is self-published by the market, this `relative-strength` strategy can be used in any market (US-Equities, Indices, Futures, Forex, etc) to derive the above properties. 

## 4. Execution Strategy

Execution Strategy is another side of the same coin of a profitable trading endeavor. The following aspects are component of the execution strategy applied in this forward-test:

#### a. Least risk entries
Trade entries are to be placed at a level which represent the least risk according to the trade-setup. The `relative-strengh` strategy (property #2) helps in determining this level.

#### b. Executes trades in lower time frame than trade-setup following Supply & Demand levels
Trades in this forward-test were executed in lower timeframe (trade-strategy property #4) than the main trade-setup (i.e. in Daily Timeframe). Since the main trade-setup is a `long` trade towards the target, the majority of trades executed as part of this forward-test are `long` trades (trade-strategy property #1 ).

#### c. Utilise `short` trades for hedging purposes during pullbacks to manage risk and minimise drawdowns
Counter trades (`short` trades in this instance) can be used to hedge during pullbacks, as necessary, in order to minimise drawdowns. This hedging mechanism is a form of risk management and shall be removed as soon as the pullback ends - as shown by trade-strategy property #3 adn #5 above.

#### d. Utilise feeler trades
This forward-test uses `feeler` trades which are trades that are smaller in size (0.1 lot size in this case). These feeler trades are used in order to minimise the possible drawdown in situations where the market failed to reverse and decided to continue in the current direction.

# Part-2 Load, Prepare Data and Perform Calculations

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image
plt.style.use('Solarize_Light2')
plt.rcParams['lines.linewidth'] = 2.5
```

#### Load Data

```python
col_names = ["transaction_id", "trade_open", "position", "lot_size", "asset", "entry_price", "stop_loss", "take_profit", "trade_close", "exit_price", "commission_cost", "taxes_cost", "swap_cost", "gross_profit"]
df = pd.read_csv("demoaccount3.csv", parse_dates=[1,6], names=col_names, header=0, usecols=[0,1,2,3,4,5,8,9,10,12,13])
```

#### Clean and Format Data

```python
df.loc[70, "lot_size"] = 0
df = df.assign(
    commission_cost=df.commission_cost.fillna(0).astype(np.float64),
    swap_cost=df.swap_cost.fillna(0).astype(np.float64),
    gross_profit=df.gross_profit.str.replace(" ", "").astype(np.float64),
    lot_size=df.lot_size.astype(np.float64),
)
df = df.assign(net_profit=df[["commission_cost", "swap_cost", "gross_profit"]].sum(axis="columns"))
df_trades = df[:-1]     # df_trades contains only trades, initial_balance of 10,000 removed.
```

#### Perform Calculations

```python
# Initial Account Balance
initial_balance = (df.loc[70, "gross_profit"])
# Total Trades
total_trades = len(df_trades)

# Profit Trades
profit_trades = (df_trades
    .net_profit.ge(0)
    .sum()
                )

# Loss Trades
loss_trades = (df_trades
    .net_profit.lt(0)
    .sum()
)

# Total Net Profit
total_net_profit = (df_trades
    .net_profit
    .sum()
)

# Final Balance
final_balance = (df
    .net_profit
    .sum()
)

# Gross Profit
gross_profit = (
    df_trades[df_trades.net_profit.ge(0)]
    .net_profit
    .sum()
)

# Gross Loss
gross_loss = (
    df_trades[df_trades.net_profit.lt(0)]
    .net_profit
    .sum()
)

# Long Positions Won and Lost
long_trades = df_trades[df_trades["position"] == "buy"]
profit_from_profitable_long_trades = long_trades[long_trades.net_profit >= 0].net_profit.sum()
number_of_profitable_long_trades = len(long_trades[long_trades.net_profit >= 0])
loss_from_losing_long_trades = long_trades[long_trades.net_profit < 0].net_profit.sum()
number_of_losing_long_trades = len(long_trades[long_trades.net_profit < 0])

# Short Positions Won and Lost
short_trades = df_trades[df_trades["position"] == "sell"]
profit_from_profitable_short_trades = short_trades[short_trades.net_profit >= 0].net_profit.sum()
number_of_profitable_short_trades = len(short_trades[short_trades.net_profit > 0])
loss_from_losing_short_trades = short_trades[short_trades.net_profit < 0].net_profit.sum()
number_of_losing_short_trades = len(short_trades[short_trades.net_profit < 0])


# Profit-Factor and Expected-Payoff Calculations

total_losing_trades=df_trades[df_trades.net_profit < 0].net_profit.abs().sum()
total_profitable_trades=df_trades[df_trades.net_profit >= 0].net_profit.sum()
net_profit = total_profitable_trades - total_losing_trades
```

# Strategy Performance Calculation

### Profit Factor

Profit Factor indicates how much a strategy can earn for every dollar lost. 

The formula: `Profit Factor = [Total Profits / Total Losses]`

```python
print(f"Profit Factor: {total_profitable_trades / total_losing_trades}")
```

The strategy delivers a return of `4.79` dollar for every `1` dollar lost.

### Expected Payoff

Expected Payoff indicates the amount of money earned from all closed trade positions.

Formula:

`Expected Payoff = Net_Profit / Total_Trade_Positions`

```python
expected_payoff = net_profit / total_trades
print(f"Expected Payoff = ${expected_payoff: .2f} profit for each trade position")
```

### Absolute Drawdown

Absolute Drawdown can be calculated as the difference between the initial capital risked and a minimal point

`Absolute Drawdown = Initial Balance - Minimal Equity`

```python
minimal_equity = (df
    .sort_values("transaction_id", ascending=True)
    .net_profit
    .reset_index(drop=True)
    .cumsum()
    .min()
)
```

```python
print(f"Absolute Drawdown = {initial_balance - minimal_equity:.2f}")
```

Absolute Drawdown is `$ 79.51`, which is less than `1%` of the initial account balance.
